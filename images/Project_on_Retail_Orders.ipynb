```python
#read data from the file and handle null valus 
import pandas as pd
data=pd.read_csv("orders.csv",na_values=['Not Available', 'unknown'])

data.head(20) 

#checkking unique name in the column shipmode 

data["Ship Mode"].unique()

#We passed  'Not Available', 'unknown' in to the read_csv method to consider this as NaN
data.head(3) 
```

```python
#Count null values 
null_values = data.isnull().sum()
print(null_values)

#Get rows where all values are null:
#Use isnull().all() to filter rows where all columns are null.
#rows_with_null = data[data.isnull().any(axis=1)]
#print(rows_with_null)

#Remove null values using dropna() -will remove all null values 
removeNull=data.dropna()
print(removeNull)

#null_values = removeNull.isnull().sum()
#print(null_values)
```

```python
#Rename column.make them in lower case and replace space with underscore 
#data.rename(columns={"Order Date" : "order_date","City":"city"}) #  we can do like this but it will take more time 

data.columns=data.columns.str.lower() #converts column names into string object and conver to lower case

data.columns #checking all are in lower case

data.columns=data.columns.str.replace(" ","_")  # replacing _ with the help of string replace method

data.columns
data.head()
```

```python
#Derive new column discount , sales_price and profit
#In discount column we have how much money discounted , we converted % to rupees
#data.drop(columns=['profit'], inplace=True )

data["discount"]=data["list_price"]*data["discount_percent"]/100

data["sale_price"]=data["list_price"]-data["discount"]

data['profit']=data['sale_price']-data['cost_price']
data.head()
```

```python
#convert order_date datatype from object to datetime
data['order_date']=pd.to_datetime(data['order_date'], format='%Y-%m-%d')
data.dtypes
```

```python
#Dropping unwanted columns 
data.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
```
